```python
import pandas as pd
import yfinance as yf
import datetime as datetime
import time
```

```python
df = pd.read_csv('./data/house_senate_years_donations.csv')
```

```python
max(df['transaction_date'])
```

```python
df = df[df['transaction_date'] < '2020-11-25']
```

```python
df.loc[df['ticker'] == 'TMUSR', 'ticker'] = 'TMUS'
df.loc[df['ticker'] == 'CCXX', 'ticker'] = 'CCX'
df.loc[df['ticker'] == 'GLIBA', 'ticker'] = 'GLIBP'
df.loc[df['ticker'] == 'LBDAV', 'ticker'] = 'LBRDA'
df.loc[df['ticker'] == 'SERV', 'ticker'] = 'TMX'
df.loc[df['ticker'] == 'SNE', 'ticker'] = 'SONY'
df.loc[df['ticker'] == 'CTL', 'ticker'] = 'LUMN'
df.loc[df['ticker'] == 'LUMNT', 'ticker'] = 'CTLT'
df.loc[df['ticker'] == 'POL', 'ticker'] = 'AVNT'
df.loc[df['ticker'] == 'APPL', 'ticker'] = 'AAPL'
```

```python
df['ticker'] = [tick.replace('CCXX', 'CCX') for tick in df['ticker']]
df['ticker'] = [tick.replace('APPL', 'AAPL') for tick in df['ticker']]
df['ticker'] = [tick.replace('GLIBA', 'GLIBP') for tick in df['ticker']]
df['ticker'] = [tick.replace('LBDAV', 'LBRDA') for tick in df['ticker']]
df['ticker'] = [tick.replace('SERV', 'TMX') for tick in df['ticker']]
df['ticker'] = [tick.replace('SNE', 'SONY') for tick in df['ticker']]
df['ticker'] = [tick.replace('CTL', 'LUMN') for tick in df['ticker']]
df['ticker'] = [tick.replace('LUMNT', 'CTLT') for tick in df['ticker']]
df['ticker'] = [tick.replace('POL', 'AVNT') for tick in df['ticker']]
df['ticker'] = [tick.replace('TMUSR', 'TMUS') for tick in df['ticker']]
```

```python
df = df[df['ticker']!= 'WMGI']
df = df[df['ticker'] != 'VSLR']
df = df[df['ticker'] != 'CTRCF']
df = df[df['ticker'] != 'DEACU']
df = df[df['ticker'] != 'RF$A']
df = df[df['ticker'] != 'TDDXX']
df = df[df['ticker'] != 'CHL']
df = df[df['ticker'] != 'AGN']
df = df[df['ticker']!= 'SNOXX']
df = df[df['ticker'] != 'AMTD']
df = df[df['ticker'] != 'WPX']
df = df[df['ticker'] != 'MSBHY']
df = df[df['ticker'] != 'VRTU']
```

```python
df.reset_index(inplace = True)
df.drop(columns = 'index', inplace = True)
```

```python
df.shape
```

```python
delta = datetime.timedelta(days=180)
```

```python
start = []
day30 = []
day60 = []
day90 = []
day180 = []
sp_start = []
sp_30 = []
sp_60 = []
sp_90 = []
sp_180 = []
for i, row in df.iterrows():
    stock = yf.download(row['ticker'] + ' ^GSPC', row['transaction_date'],
            str((datetime.datetime.strptime(row['transaction_date'],"%Y-%m-%d") + delta))[:10])
    start.append(stock['Close'][row['ticker']][0])
    day30.append(stock['Close'][row['ticker']][21])
    day60.append(stock['Close'][row['ticker']][42])
    day90.append(stock['Close'][row['ticker']][65])
    day180.append(stock['Close'][row['ticker']][-1])
    sp_start.append(stock['Close']['^GSPC'][0])
    sp_30.append(stock['Close']['^GSPC'][21])
    sp_60.append(stock['Close']['^GSPC'][42])
    sp_90.append(stock['Close']['^GSPC'][65])
    sp_180.append(stock['Close']['^GSPC'][-1])
    print(i)
    time.sleep(1)
```

```python
yf.download(df['ticker'][0], df['transaction_date'][0],
            str((datetime.datetime.strptime(df['transaction_date'][0],"%Y-%m-%d") + delta))[:10])
```

```python
df['start'] = start
df['day_30'] = day30
df['day_60'] = day60
df['day_90'] = day90
df['day_180'] = day180
df['sp_start'] = sp_start
df['sp_30'] = sp_30
df['sp_60'] = sp_60
df['sp_90'] = sp_90
df['sp_180'] = sp_180
```

```python
df.isna().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.to_csv('./data/final_dataset.csv', index=False)
```
