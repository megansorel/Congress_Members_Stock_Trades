# Visualizations

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Import Data
df = pd.read_csv('./data/stock_percentage_difference.csv')
```

30 Day to 180 Day performance

```python
# Create Percentage difference arrays for graph
Cong = (df['day_180']/df['day_30'])-1
SP = (df['sp_180']/df['sp_30'])-1

# Create graph
fig, ax = plt.subplots(1,1, figsize=(12,12))
ax.set_title('Day 30 to Day 180 Comparison of Congression Trades vs. S&P 500', loc='center', size=35, pad=20)
ax.axhline(0, color='white', label='No Change')
ax.scatter(df['level_0'], Cong, label='Congress Trades', alpha=.5)
ax.scatter(df['level_0'], SP, label='S&P', alpha=.5)
ax.legend(loc = (1.05, .4), prop={'size':25}, facecolor='black', labelcolor='white')
ax.set_ylabel('Percentage Change', size=35);
```
