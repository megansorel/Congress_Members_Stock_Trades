```python
people = []
senate_years = []
house_years = []
party = []
state = []
total_years = []
#get all members 


for i in range (1, 11):
    url = 'https://www.congress.gov/members?searchResultViewType=compact&pageSize=250&page={}'.format(i)
    res = requests.get(url)
    print(res.status_code)
    soup = BeautifulSoup(res.content, 'lxml')
    for listing in soup.find_all('li', {'class': 'compact'}):
            # get member name under result heading

        for name in listing.find_all('span', {'class': 'result-heading'}):
            people.append(name.text)


        for info in listing.find_all('span', {'class': 'result-item'}):
                #get strong and span, filter by strong 
            for strong in info.find_all('strong'):
                if strong.text == 'State:':
                    state.append([span.text for span in info.find_all('span')])
                if strong.text == 'Party:':
                    party.append([span.text for span in info.find_all('span')])
        for years in listing.find_all('ul', {'class': 'member-served'}):
                total_years.append(years.text)
```

```python
people = [p.split(' ') for p in people]
```

```python
last_title = [p[0] for p in people]
people = [p[1:] for p in people]
```

```python
last_title
```

```python
first_name = [p[1] for p in people]

last_name = [p[0].strip(',') for p in people]

middle = [p[2] if len(p) == 3 else '' for p in people]

names = [' '.join(person) for person in zip(first_name, last_name)]

full_names = [', '.join(person) for person in zip(names, middle)]

full_names
```

```python
total_years = [string.strip('\n') for string in total_years]
total_years
```

```python
lst = []
for row in total_years:
    row = row.split(' ')
    lst.append(row)
```

```python
senate_years = []
house_years = []
for row in lst:
    if 'Senate:' in row:
        senate_years.append(row[1])
    else: 
        senate_years.append(0)
```

```python
for row in lst:
    if 'House:' in row:
        index = row.index('House:') + 1
        house_years.append(row[index])
    else: 
        house_years.append(0)
```
