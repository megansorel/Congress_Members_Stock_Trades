```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
people = []
senate_years = []
house_years = []
party = []
state = []
total_years = []
#get all members 


for i in range (1, 11):
    url = 'https://www.congress.gov/members?searchResultViewType=compact&pageSize=250&page={}'.format(i)
    res = requests.get(url)
    print(res.status_code)
    soup = BeautifulSoup(res.content, 'lxml')
    for listing in soup.find_all('li', {'class': 'compact'}):
            # get member name under result heading

        for name in listing.find_all('span', {'class': 'result-heading'}):
            people.append(name.text)


        for info in listing.find_all('span', {'class': 'result-item'}):
                #get strong and span, filter by strong 
            for strong in info.find_all('strong'):
                if strong.text == 'State:':
                    state.append([span.text for span in info.find_all('span')])
                if strong.text == 'Party:':
                    party.append([span.text for span in info.find_all('span')])
        for years in listing.find_all('ul', {'class': 'member-served'}):
                total_years.append(years.text)
```

```python
len(people)
```

```python
people = [p.split(' ') for p in people]
```

```python
last_title = [p[0] for p in people]
people = [p[1:] for p in people]
```

```python
first_name = [p[1] for p in people]
last_name = [p[0].strip(',') for p in people]
middle = [p[2] if len(p) == 3 else '' for p in people]
names = [' '.join(person) for person in zip(first_name, last_name)]
full_names = [', '.join(person) for person in zip(names, middle)]
```

```python
total_years = [string.strip('\n') for string in total_years]
for row in total_years:
    print(row, '\n')
```

```python
senate = []
house = []

for row in total_years:
    if 'Senate:' in row and 'House:' in row:
        row = row.split('House:')
        senate.append(row[0])
    elif 'Senate:' in row and 'House:' not in row:
        senate.append(row)
    else:
        senate.append('None')        
        
for row in total_years:
    if 'Senate:' in row and 'House:' in row:
        row = row.split('House:')
        house.append(row[1])
    elif 'House:' in row and 'Senate:' not in row:
        house.append(row)
    else:
        house.append('None')


senate = [s.strip('Senate:').strip(' ') for s in senate]
house = [h.strip('House:').strip(' ') for h in house]
```

# Cleaning the data 

```python
df = pd.read_csv('./data/congresspeople_years_party_state.csv')
```

```python
df.head()
```

```python
df['people'] = [x.strip(',') if x[-1]==',' else x for x in df['people']]
```

```python
df
```

```python
years = []
for row in df['senate_years']:
    years.append(row.replace('Present', '2021'))
df['senate_present_updated'] = years
```

```python
house_lst = []
for row in df['house_years']:
    house_lst.append(row.replace('Present', '2021'))
df['house_present_updated'] = house_lst
```

```python
#getting all senate years

total_years = []
for row in df['senate_present_updated']:
    if row == 'No':
        total_years.append(0)
    elif ',' in row:
            row = row.split(',')
            first_term = row[0]
            second_term = row[-1]

            #getting start and end of first terms
            first_term = first_term.split('-')
            first_start = first_term[0]
            first_end = first_term[-1]

            #getting start and end of second term
            second_term = second_term.split('-')
            second_start = second_term[0]
            second_end = second_term[-1]

            #subtracting start from end
            first_years = int(first_end) - int(first_start)
            second_years = int(second_end) - int(second_start)

            #adding term years together for full service
            total_years.append(first_years + second_years)

    else:
        term = row.split('-')
        start = term[0]
        end = term[-1]
        total_years.append(int(end) - int(start))
```

```python
total_house_years = []
for row in df['house_present_updated']:
    if row == 'Non':
        total_house_years.append(0)
    elif ',' in row:
            row = row.split(',')
            first_term = row[0]
            second_term = row[-1]

            #getting start and end of first terms
            first_term = first_term.split('-')
            first_start = first_term[0]
            first_end = first_term[-1]

            #getting start and end of second term
            second_term = second_term.split('-')
            second_start = second_term[0]
            second_end = second_term[-1]

            #subtracting start from end
            first_years = int(first_end) - int(first_start)
            second_years = int(second_end) - int(second_start)

            #adding term years together for full service
            total_house_years.append(first_years + second_years)

    else:
        term = row.split('-')
        start = term[0]
        end = term[-1]
        total_house_years.append(int(end) - int(start))
```

```python
df['total_senate_years'] = total_years
```

```python
df['total_house_years'] = total_house_years
```

```python
df.head()
```

```python
final = df.drop(columns = ['senate_years', 'house_years', 'senate_present_updated', 'house_present_updated'])
```

```python
final['total_years'] = df['total_senate_years'] + df['total_house_years']
```
